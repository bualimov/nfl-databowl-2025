```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print('imports done.')
```

```python
# reading in files (weeks)
wk1 = pd.read_csv("tracking_week_1.csv", na_values=["NA"])
wk2 = pd.read_csv("tracking_week_2.csv", na_values=["NA"])
wk3 = pd.read_csv("tracking_week_3.csv", na_values=["NA"])
wk4 = pd.read_csv("tracking_week_4.csv", na_values=["NA"])
wk5 = pd.read_csv("tracking_week_5.csv", na_values=["NA"])
wk6 = pd.read_csv("tracking_week_6.csv", na_values=["NA"])
wk7 = pd.read_csv("tracking_week_7.csv", na_values=["NA"])
wk8 = pd.read_csv("tracking_week_8.csv", na_values=["NA"])
wk9 = pd.read_csv("tracking_week_9.csv", na_values=["NA"])

# other stuff
player_play = pd.read_csv("player_play.csv", na_values=["NA"])
games = pd.read_csv("games.csv", na_values=["NA"])
players = pd.read_csv("players.csv", na_values=["NA"])
plays = pd.read_csv("plays.csv", na_values=["NA"])

print('charts read in.')
```

```python
# attempt to combine all into one df

# declare list of all weeks
weeks = [wk1, wk2, wk3, wk4, wk5, wk6, wk7, wk8, wk9]

# concat
df = pd.concat(weeks, ignore_index=True)

# successfully combined
# time: 51.4s
print('big data set (df) formed, with dimensions: ', df.shape)

#view
df.head(5)
```

```python
# big df to csv, to save in folder
df.to_csv("combined_weeks.csv", index=False)
print('succesfully saved big df as csv to folder.')

# time: 3m 43.3s
```

```python
# now combining df + player_play
df1 = pd.merge(df, player_play, on=["gameId", "playId", "nflId"], how="inner")

# it worked
print('big data set (df1) formed, with dimensions: ', df1.shape)
df1.head(5)
```
